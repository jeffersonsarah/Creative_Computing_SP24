Python provide two loop categories:

* Condition control loops: True/false conditions to control each loop. --------> `while`
* Count control loops: repeat a task or action a specific number of times.-----> `for` 

## While Loop: Condition-controlled loop
A condition-controlled loop causes a statement or set of statements to 
repeat as long as a condition is true.



```python
x = 0
while x < 3:
    print(f'I am inside the loop {x+1}')
    x += 1
print(f'I am outside of the loop')
```

```python
count = 1  #initialize your conditional variable ALWAYS
control = int(input(f'How many loops do you wish to execute? '))
while count <= control:   #conditional
    print(count)     #action 1
    count = count + 1 #action 2
#prints from 0 to 4
```
