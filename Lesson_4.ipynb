## Comparison operators
!


```python
print(3 > 2)     # T
print(3 >= 2)    # T
print(3 < 2)     # F
print(2 < 3)     # T
print(2 <= 3)    # T
print(3 == 2)    # F
print(3 != 2)    # T
print('mango' == 'avocado')  #F
print('mango' != 'avocado')  # T
print('mango' < 'avocado')   # F
print('Mark' != 'Mary')      # T
print('Mark' == 'Mary')      # F
print('tomato' == 'potato')  # F
print('python' > 'dragon')   # T
```

Python offers reserved words as comparison operators:

* in: Returns True if the queried list contains a certain item(x in y)
* not in: Returns True if the queried list doesn't have a certain item(x in y)

```python
print('S in Seah', 'S' in 'Seah') # T
print('X in Seah', 'X' in 'Seah') # F
print('end' in 'Friend') # T
print('a in an:', 'a' in 'an')      # T
```

## Logical Operators
Logical operators are useful as combinations of comparison operators. Using their truth tables, their output is boolean.
!

```python
print(3 > 2 and 4 > 3) # T AND T = T
print(3 > 2 and 4 < 3) # T AND F = F
print(3 < 2 and 4 < 3) # F AND F = F
print(3 > 2 or 4 > 3)  # T OR T = T
print(3 > 2 or 4 < 3)  # T OR F = T

print(not 3 > 2)     # F
print(not True)      # F
print(not False)     # T
print(not not True)  # T
print(not not False) # F
```

# Decision Structures and Conditionals

By default, statements in Python script are executed sequentially from top to bottom. If the processing logic require so, the sequential flow of execution can be altered in two way:

* Conditional execution: a block of one or more statements will be executed based on a true or false decision.
* Repetitive execution: a block of one or more statements will be repetitively executed as long as a certain expression is true.

### Example 1

Dr. Seah teaches a Creative Computing course and her students are required to take three tests. She wants 
to write a program that her students can use to calculate their average test score. She also 
wants the program to congratulate the student if they pass the course (above 59%) or encourage them to try harder if they fail (below 60%). 

```python
#write code here
#three inputs from user. test1, test2, test3
test1 = int(input('Enter Test 1: '))
test2 = int(input('Enter Test 2: '))
test3 = int(input('Enter Test 3: '))
#calculate avergae grade
average = (test1 + test2 + test3)/3

#compare average with passing grade
if average >= 60:
    print(f'Your grade is {average:.0f}%. CONGRATULATIONS!') #action for true
else:
    print(f'Your grade is {average:.0f}%. Please try harder next time.')

```

### Example 2

Tyler owns an auto repair business and has several employees. If any employee works over 
40 hours in a week, he pays them 1.5 times their regular hourly pay rate for all hours over 
40. He has asked you to design a simple payroll program that calculates an employees 
gross pay, including any overtime wages. The program should get employee's name, number of hours worked, and normal hourly pay rate. A final sentence of employee's name and gross pay is expected.

```python
#write code here
#inputs: name, hours_worked, pay_rate
name = input("Enter employee's name: ")
hours_worked = float(input("Enter number of hours worked this week: "))
pay_rate = float(input("Enter pay rate (do not include $ sign: "))
salary = 0

#compare: if hours_worked > 40, then extra hours have a 1.5 pay rate
if hours_worked > 40:
    #calculate first 40 hours normally
    salary = 40* pay_rate + (hours_worked - 40) * pay_rate * 1.5
else:
    salary = hours_worked * pay_rate
print(f"Employee's name: {name} \nGross pay: ${salary:.2f}.")

# 45 hours, 10$/hour
# salary = 40*10 + 5*10*1.5 =475$
# false case. 30 hours --> salary = 30*10 = 300$
```

### Example 4
Dr. Suarez teaches a literature class and uses the following 10-point grading scale for all of 
his exams:
* A --> 90 or above
* B --> 80-89
* C --> 70-79
* D --> 60-69
* F --> Below 60

He has asked you to write a program that will allow a student to enter a test score and then 
display the grade for that score. Here is the algorithm that you will use:
1.  Ask the user to enter a test score.
2.  Display the letter grade using a sentence.


```python
#write code here
#inputs: grade
grade = float(input("Enter grade: "))

if grade >= 90:
    print("Your grade is A")
elif grade >= 80 and grade < 90:
    print("Your grade is B")
elif grade >= 70 and grade < 80:
    print("Your grade is C")
elif grade >= 60 and grade < 70:
    print("Your grade is D")
else:
    print("Your grade is F")
```
